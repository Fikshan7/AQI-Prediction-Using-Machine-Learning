**Handling Missing Values**

```python
#Converting All the Missing Values from -200 to NaN
air_quality_data=air_quality_data.replace(to_replace=-200,value= np.nan)
```

```python
#Checking Missing Values fromm the dataframe
air_quality_data.isnull().sum()
```

```python
air_quality_data.iloc[:, 2:].mean()
```

```python
#Replace the Missing value with the mean of the mean value
air_quality_data=air_quality_data.fillna(air_quality_data.iloc[:, 2:].mean())
```

**Forecasting With FB Prophet Model**

FB Prophet Documentation:https://facebook.github.io/prophet/docs/quick_start.html#python-api

```python
date_info=pd.to_datetime(air_quality_data['Date'], format="%d/%m/%Y")
print(date_info)
```

```python
time_info=air_quality_data['Time']
print(time_info)
```

```python
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(time_info))
print(type(date_info))
```

```python
#combining Two Series to panda Dataframe
date_time=pd.concat([date_info, time_info], axis=1)
```

```python
date_time.head()
```

```python
date_time.shape
```

```python
#combining Date and Time
date_time['ds']= date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

We have to change ds datatype from object to date & time datatype

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

**Training the Prophet Model**

```python
model=Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
